### Data Description

- Customer - Customer ID

- ST - State where customers live

- Gender - Gender of the customer

- Education - Background education of customers 

- Customer Lifetime Value - Customer lifetime value(CLV) is the total revenue the client will derive from their entire relationship with a customer. In other words, is the predicted or calculated value of a customer over their entire duration as a policyholder with the insurance company. It is an estimation of the net profit that the insurance company expects to generate from a customer throughout their relationship with the company. Customer Lifetime Value takes into account factors such as the duration of the customer's policy, premium payments, claim history, renewal likelihood, and potential additional services or products the customer may purchase. It helps insurers assess the long-term profitability and value associated with retaining a particular customer.

- Income - Customers income

- Monthly Premium Auto - Amount of money the customer pays on a monthly basis as a premium for their auto insurance coverage. It represents the recurring cost that the insured person must pay to maintain their insurance policy and receive coverage for potential damages, accidents, or other covered events related to their vehicle.

- Number of Open Complaints - Number of complaints the customer opened

- Policy Type - There are three type of policies in car insurance (Corporate Auto, Personal Auto, and Special Auto)

- Vehicle Class - Type of vehicle classes that customers have Two-Door Car, Four-Door Car SUV, Luxury SUV, Sports Car, and Luxury Car

- Total Claim Amount - the sum of all claims made by the customer. It represents the total monetary value of all approved claims for incidents such as accidents, theft, vandalism, or other covered events.

```python
#Read the csv file and show type

file1 = pd.read_csv('file1.csv')
print("The variable type of file1 is: ", type(file1))
print()

#Show first 5 rows
print("These are the first five rows of the DataFrame: ")
display(file1.head())


#Show last 5 rows
print("These are the last five rows of the DataFrame: ")
display(file1.tail())


#Size of DataFrame
print("This is the size of the DataFrame: ",file1.shape)

#Row and Columns
print("\nThe DataFrame has {} rows and {} columns".format(file1.shape[0], file1.shape[1]))
dtypes = file1.dtypes


#Show data types
print("\nData Types of Columns:")
print(dtypes)
```

To deal with the missing values, there are several possible strategies. Among the most basic ones are:

Drop the column that contain missing values:

We might loose relevant information for our analysis and future predictions.

Drop the rows that contain missing values:

We might reduce a lot the amount of rows in our DataFrame and compromise any statistical information that we can compute.

Replace them by another value:

using "background knowledge" (we know the value by any reason)
computed from the same column that has the missing values
computed from other columns of the same row
computed using other more sophisticated methods
If we decide to drop the rows that contain the missing values, it's important to compute first the percentage of missing values in the column.

```python
#Idenify the unique vales for each column

unique_values = file1.nunique()
print("\nNumber of uniques values for columns:")
print(unique_values)

#Identify columns that appear to be categorical using list, using limit of 12 for brevity.

categorical_columns = file1.columns[file1.nunique() < 12]

# Gets unique values from list

categorical_values = {col: file1[col].unique() for col in categorical_columns}

print("Unique Values for Categorical Columns:")
for col, values in categorical_values.items():
    display(f"{col}: {values}")
    
# Values for numerical columns
numerical_columns = file1.select_dtypes(include=['float64', 'int64']).columns

# Get range min and max for each numerical column
numerical_range = {col: (file1[col].min(), file1[col].max()) for col in numerical_columns}

print("\nRange of Values for Numerical Columns:")
for col, values in numerical_range.items():
    print(f"{col}: {values}")
```

Observations:

The min of the lowest claim amount also seems like an outlier.

```python
# Statistics for numerical columns
descriptive_stats = file1[numerical_columns].describe()
print("\nSummary Statistics for Numerical Columns:")
display(descriptive_stats)

# Statistics for categorical columns
categorical_stats = pd.DataFrame({col: file1[col].value_counts() for col in categorical_columns}).transpose()
print("\nSummary Statistics for Categorical Columns:")
display(categorical_stats)
```
